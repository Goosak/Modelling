```python
import pandas as pd
#for visualization
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
#for model
from sklearn.tree import DecisionTreeClassifier #A decision tree classifier.
from sklearn.preprocessing import LabelEncoder #for converting strings to fit form
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split #Split arrays or matrices into random train and test subsets
```

```python
def convert(data, columns):
    number=LabelEncoder()
    for c in columns:
        data[c] = number.fit_transform(data[c])
    return data
```

```python
credit = pd.read_csv("datasets\\CreditSample.txt", delimiter="\t", header=0, index_col=None, encoding='cp1251')
credit.head()
```

```python
columns = ['Давать кредит', 'Адрес', 'Код', 'Название организации', 'ФИО', '№ Паспорта']
X=credit[credit.columns.difference(columns)]
Y=credit[columns[0]].to_frame()
X.head()
```

```python
columns_X=X.columns.tolist()
columns_Y=Y.columns.tolist()
X=convert(X, columns_X)
Y=convert(Y, columns_Y)
```

```python
Y.head()
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.25, random_state=1)
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```
