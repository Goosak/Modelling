# Подготовка данных

```python
sport_train=sport_train[(sport_train.Decision_Making<=100)&(sport_train.Decision_Making>=3)]
sport_score=sport_score[(sport_score.Decision_Making<=100)&(sport_score.Decision_Making>=3)]
sport_train = convert(sport_train, ["Prime_Sport"])
```

```python
x = sport_train.iloc[:, :8]
y = sport_train.iloc[:, 8]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

# Обучение

```python
classifier = RandomForestClassifier(max_depth=2)

classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

# Оценка модели

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy: ' + str(accuracy_score(y_test, y_pred)))
```
