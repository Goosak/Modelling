```python
import numpy as np
import pandas as pd
from convert import convert
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
path="datasets\\"
dataset="CreditSample.txt"

credit = pd.read_csv(path+dataset, sep="\t", encoding="cp1251")
print(credit.columns)
credit.head()
```

```python
cols=['Код', '№ Паспорта', 'Основное направление расходов', 'Название организации', 'Отраслевая принадлежность предприятия',
     'Адрес', 'ФИО']

credit=credit.drop(cols, axis=1)
credit.columns
```

```python
credit = convert(credit, credit.columns)
```

```python
credit.head()
```

```python
x = credit.iloc[:, credit.columns!="Давать кредит"]
y = credit.iloc[:, credit.columns=="Давать кредит"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
clf = LogisticRegression(random_state = 0).fit(x_train, y_train)
predict = clf.predict(x_test)
```

```python
confusion_matrix(y_test, predict)
```

```python
clf.score(x, y)
```
